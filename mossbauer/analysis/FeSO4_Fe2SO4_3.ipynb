```python
%pylab inline
from __future__ import division, print_function
import jlab as j
import cPickle
import pygsl as g; g.import_all()

class Bundle: pass

cal = cPickle.load(open('calibration.p', 'rb'))
misc = cPickle.load(open('misc.p', 'rb'))

fe4 = Bundle()
fe43 = Bundle()

fe4.raw = genfromtxt('../FeSO4_Fe2SO4_3/02-15-FeSO4.csv', delimiter=',', skip_header=17)
fe43.raw = genfromtxt('../FeSO4_Fe2SO4_3/02-15-Fe2SO4_3.csv', delimiter=',', skip_header=17)
```

```python
# Raw Data
figure()
plot(fe4.raw[:,0],fe4.raw[:,2],'b.')
plot(fe43.raw[:,0],fe43.raw[:,2],'g.')
```

```python
# bin data
bin_cutoff = 200
bins = linspace(fe4.raw[bin_cutoff,0], fe4.raw[-1,0], 250)

def bin_bundle(bundle, bins):
    digitized = numpy.digitize(bundle.raw[:,0], bins)
    bundle.x = bins[:-1]
    bundle.y = array([j.weighted_mean(bundle.raw[:,2][digitized == i], sqrt(bundle.raw[:,2][digitized == i])) for i in range(1, len(bins))])
    bundle.ye = bundle.y[:,1]
    bundle.y = bundle.y[:,0]
    if 0 in bundle.ye:
        print('There is a 0 in bin error at:', where(bundle.ye == 0))
        
bin_bundle(fe4, bins)
bin_bundle(fe43, bins)
fe4.vel = cal.vel.vel_per_channel[[int(x) for x in fe4.x]]
fe43.vel = cal.vel.vel_per_channel[[int(x) for x in fe43.x]]
```

```python
figure()
errorbar(fe4.x, fe4.y, fe4.ye,fmt='.')
errorbar(fe43.x, fe43.y, fe43.ye,fmt='.')
```
