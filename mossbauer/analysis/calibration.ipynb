```python
%pylab inline
from __future__ import division, print_function
import jlab as j
import cPickle
import pygsl as g; g.import_all()

class Bundle(): pass

# import data
misc = Bundle()
misc.E0 = 14.4 #keV
misc.lam = 6.328e-7 # wavelength of laser [m]
misc.DT = 100.0e-6 # dwell time [sec]
misc.N = 4000 # passes [1]

vel = Bundle()
vel.raw = genfromtxt('../pos_cal_1/02-14-V_2.csv', delimiter=',', skip_header=17)

fe = Bundle()
fe.raw = sum(genfromtxt('../pos_cal_1/'+x+'.csv', delimiter=',', skip_header=17) for x in ['02-14-6_1', '02-14-6_2'])
```

```python
bin_cutoff = 200
# Raw Data Figure
figure()
plot(range(len(fe.raw[:,2])), fe.raw[:,2],'.')
axvline(bin_cutoff, -10, 10, c='r')
xlabel('Channels [1]')
ylabel('Counts [1]')
```

```python
# Bin Data
bins = linspace(fe.raw[bin_cutoff,0]/2, fe.raw[-1,0]/2, 250)
digitized = numpy.digitize(fe.raw[:,0]/2, bins)
fe.x = bins[:-1]
fe.y = array([j.weighted_mean(fe.raw[:,2][digitized == i], sqrt(fe.raw[:,2][digitized == i])) for i in range(1, len(bins))])
fe.ye = fe.y[:,1]
fe.y = fe.y[:,0]

if 0 in fe.ye:
    print('There is a 0 in bin error at:', where(fe.ye == 0))

figure()
errorbar(fe.x, fe.y, fe.ye, fmt='.')
xlabel('Channels [1]')
ylabel('Counts [1]')
```

```python
# Vel Raw Data
figure()
errorbar(vel.raw[:,0], vel.raw[:,2], sqrt(vel.raw[:,2]),fmt='k.')
axvline(bin_cutoff, -10, 10, c='r')
xlabel('Channels [1]')
ylabel('Counts [1]')
grid()
```

```python
# Fit 'V' Curve to Velocity Data
def speed_func(v, a):
    a1, b1, a2, b2 = a
    min_point = (b2-b1)/(a1-a2)
    if isscalar(v):
        if v <= min_point:
            return a1*v+b1
        else:
            return a2*v+b2
    return concatenate([a1*v[v <= min_point]+b1, a2*v[v > min_point]+b2], axis=1)

def vel_func(v, a):
    # TODO: This forces continuous function; may not be ideal.
    a1, b1, a2, b2 = a
    min_point = (b2-b1)/(a1-a2)
    if isscalar(v):
        if v <= min_point:
            return -a1*(v-min_point)
        else:
            return a2*(v-min_point)
    return concatenate([-a1*(v[v <= min_point]-min_point), a2*(v[v > min_point]-min_point)], axis=1)

vel.a0 = array([-15000/800., 20000., 15000/800., -20000.])
vel.a, vel.aerr, vel.chisq, vel.yfit = j.levmar(vel.raw[bin_cutoff:,0], vel.raw[bin_cutoff:,2], sqrt(vel.raw[bin_cutoff:,2]), speed_func, vel.a0, 
        print_dev=True)[:-1]
```

```python
figure()
plot(vel.raw[bin_cutoff:,0], speed_func(vel.raw[bin_cutoff:,0], vel.a0), 'g')
errorbar(vel.raw[:,0], vel.raw[:,2], sqrt(vel.raw[:,2]),fmt='k.')
plot(vel.raw[bin_cutoff:,0], vel.yfit, 'r')
plot(vel.raw[bin_cutoff:,0], vel_func(vel.raw[bin_cutoff:,0], vel.a))
print('$chi^2$: {}'.format(vel.chisq))
```

```python
# Compute Velocity of each channel

# Linear Velocity
vel.vel_per_channel = (vel_func(vel.raw[:,0], vel.a))*misc.lam/(2*misc.N*misc.DT)
vel.vel_per_channel_error = misc.lam/(2*misc.N*misc.DT)*sqrt(vel.raw[:,2]) # TODO: Correct error propogation
fe.vel = vel.vel_per_channel[[int(x) for x in fe.x]]

if False: # Non-linear Velocity
    vel.vel_per_channel = (vel.raw[:,2])*misc.lam/(2*misc.N*misc.DT)
    vel.vel_per_channel[:argmin(vel.vel_per_channel)] *= -1
    vel.vel_per_channel_error = misc.lam/(2*misc.N*misc.DT)*sqrt(vel.raw[:,2])
    fe.vel = vel.vel_per_channel[[int(x) for x in fe.x]]
```

```python
# Velocity per channel plot
figure()
errorbar(range(len(vel.vel_per_channel)), vel.vel_per_channel, vel.vel_per_channel_error, fmt='k.')
grid()
ylabel('Velocity[m/s]')
xlabel('Channel [1]')
```

```python
# Photon Energy
figure()
E_per_channel = misc.E0*(1+vel.vel_per_channel/g.const.speed_of_light)
E_per_channel_error = misc.E0*vel.vel_per_channel_error/g.const.speed_of_light
#errorbar(range(len(E_per_channel)), E_per_channel,E_per_channel_error, fmt='k.')
```

```python
# Manual Intervention for Approximate location of peaks
approx = Bundle()
approx.peak_channel = array([528, 742, 955, 1120, 1352, 1586])
approx.peak_max_count = array([120,120,120,120,120,120])/6
approx.peak_mins = array([71,68,84,83,70,74]) - 5*int(mean(approx.peak_max_count))
approx.peak_fwhm = array([50,50,50,50,50,50])

def a_peak(V, a):
    ''' See Eq. 6, lab guide '''
    C0, B, DE, E0, T = a
    B /= 1e22
    DE /= 1e10
    T /= 1e10
    return C0*(1-B/((E0*V/g.const.speed_of_light - DE)**2 + T**2))

def six_peaks(V, a):
    x = zeros(V.shape)
    for i in range(6):
        x += a_peak(V, array([a[i], a[i+6], a[i+12], a[-1], a[i+18]]))
    return x

DE = vel.vel_per_channel[approx.peak_channel.astype(int)]/g.const.speed_of_light*misc.E0
T = abs(vel.vel_per_channel[(approx.peak_channel+approx.peak_fwhm/2).astype(int)]-
        vel.vel_per_channel[(approx.peak_channel-approx.peak_fwhm/2).astype(int)])*misc.E0/(2*g.const.speed_of_light)
B = -(approx.peak_mins/approx.peak_max_count-1)*T**2
fe.a0 = concatenate([approx.peak_max_count, B*1e22, DE*1e10, T*1e10, [misc.E0]], axis=1)


figure()
errorbar(fe.vel, fe.y, fe.ye, fmt='.')
plot(vel.vel_per_channel, six_peaks(vel.vel_per_channel, fe.a0), 'g')
```

```python
# Conduct six peak fit (slow)
fe.a, fe.aerr, fe.chisq, fe.yfit = j.levmar(fe.vel, fe.y, fe.ye, six_peaks, fe.a0, print_dev=True)[:-1]
```

```python
# Plot Fit Results
print('$chi^sq$: {}'.format(fe.chisq))
figure()
errorbar(fe.vel, fe.y, fe.ye, fmt='.')
plot(fe.vel, fe.yfit, 'r')
xlabel('Velocity [1]')
ylabel('Counts [1]')
```
