```python
%pylab inline
from __future__ import division, print_function
import jlab as j
import cPickle
import pygsl as g; g.import_all()

class Bundle: pass

cal = cPickle.load(open('calibration.p', 'rb'))
misc = cPickle.load(open('misc.p', 'rb'))

fe = Bundle()

fe.raw = genfromtxt('../Fe3O4/02-15-Fe3O4.csv', delimiter=',', skip_header=17)
```

```python
# Raw Data
figure()
plot(fe.raw[:,0],fe.raw[:,2],'b.')
```

```python
# bin data
bin_cutoff = 200
bins = linspace(fe.raw[bin_cutoff,0], fe.raw[-1,0], 250)

def bin_bundle(bundle, bins):
    digitized = numpy.digitize(bundle.raw[:,0], bins)
    bundle.x = bins[:-1]
    bundle.y = array([j.weighted_mean(bundle.raw[:,2][digitized == i], sqrt(bundle.raw[:,2][digitized == i])) for i in range(1, len(bins))])
    bundle.ye = bundle.y[:,1]
    bundle.y = bundle.y[:,0]
    if 0 in bundle.ye:
        print('There is a 0 in bin error at:', where(bundle.ye == 0))
        
bin_bundle(fe, bins)
fe.vel = cal.vel.vel_per_channel[[int(x) for x in fe.x]]
```

```python
figure()
errorbar(fe.x, fe.y, fe.ye,fmt='.')
```

```python
# Peak Fit Functions
def a_peak(V, a):
    ''' See Eq. 6, lab guide '''
    C0, B, DE, E0, T = a
    B /= 1e22
    DE /= 1e10
    T /= 1e10
    return C0*(1-B/((E0*V/g.const.speed_of_light - DE)**2 + T**2))

def six_peaks(V, a):
    x = zeros(V.shape)
    for i in range(6):
        x += a_peak(V, array([a[i], a[i+6], a[i+12], a[-1], a[i+18]]))
    return x

def prep_for_fit(bundle, approx, misc=misc, vel_per_channel=cal.vel.vel_per_channel):
    DE = vel_per_channel[approx.peak_channel.astype(int)]/g.const.speed_of_light*misc.E0
    T = abs(vel_per_channel[(approx.peak_channel+approx.peak_fwhm/2).astype(int)]-vel_per_channel[(approx.peak_channel-approx.peak_fwhm/2).astype(int)])*misc.E0/(2*g.const.speed_of_light)
    B = -(approx.peak_mins/approx.peak_max_count-1)*T**2
    bundle.a0 = concatenate([approx.peak_max_count, B*1e22, DE*1e10, T*1e10, [misc.E0]], axis=1)

def run_fit(fe):
    fe.a, fe.aerr, fe.chisq, fe.yfit = j.levmar(fe.vel, fe.y, fe.ye, six_peaks, fe.a0, print_dev=True)[:-1]
```

```python
# Manual Intervention for Approximate location of peaks
approx = Bundle()
approx.peak_channel = array([320, 614, 948, 1210, 1520, 1900])
approx.peak_max_count = array([1500,1500,1500,1500,1500,1500])/6
approx.peak_mins = array([1400,1400,1400,1400,1370,1350]) - 5*int(mean(approx.peak_max_count))
approx.peak_fwhm = array([50,50,50,50,50,50])
prep_for_fit(fe, approx)

figure()
errorbar(fe.vel, fe.y, fe.ye, fmt='.')
plot(cal.vel.vel_per_channel, six_peaks(cal.vel.vel_per_channel, fe.a0), 'g')
```

```python
# Conduct Six Peak Fit (slow)
run_fit(fe)
```

```python
# Plot
print('$chi^sq$: {}'.format(fe.chisq))
errorbar(fe.vel, fe.y, fe.ye, fmt='.')
plot(fe.vel, fe.yfit, 'r')
xlabel('Velocity [1]')
ylabel('Counts [1]')
```
