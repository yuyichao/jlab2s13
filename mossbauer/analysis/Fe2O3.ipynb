```python
%pylab inline
from __future__ import division, print_function
import jlab as j
import cPickle
import pygsl as g; g.import_all()

class Bundle: pass

cal = cPickle.load(open('calibration.p', 'rb'))
misc = cPickle.load(open('misc.p', 'rb'))

fe1 = Bundle()
fe2 = Bundle()

fe1.raw = genfromtxt('../Fe2O3/02-14-Fe2O3_1.csv', delimiter=',', skip_header=17)
fe2.raw = genfromtxt('../Fe2O3/02-14-Fe2O3_2.csv', delimiter=',', skip_header=17)
```

```python
# Raw Data
figure()
plot(fe1.raw[:,0],fe1.raw[:,2],'b.')
plot(fe2.raw[:,0],fe2.raw[:,2],'g.')
```

```python
# bin data
bin_cutoff = 0 # TODO: Should probably be 200 as in calibration
bins = linspace(fe1.raw[bin_cutoff,0], fe1.raw[-1,0], 250)

def bin_bundle(bundle, bins):
    digitized = numpy.digitize(bundle.raw[:,0], bins)
    bundle.x = bins[:-1]
    bundle.y = array([j.weighted_mean(bundle.raw[:,2][digitized == i], sqrt(bundle.raw[:,2][digitized == i])) for i in range(1, len(bins))])
    bundle.ye = bundle.y[:,1]
    bundle.y = bundle.y[:,0]
    if 0 in bundle.ye:
        print('There is a 0 in bin error at:', where(bundle.ye == 0))

bin_bundle(fe1, bins)
bin_bundle(fe2, bins)

fe1.vel = cal.vel.vel_per_channel[[int(x) for x in fe1.x]]
fe2.vel = cal.vel.vel_per_channel[[int(x) for x in fe2.x]]
```

```python
figure()
errorbar(fe1.x, fe1.y, fe1.ye,fmt='.')
errorbar(fe2.x, fe2.y, fe2.ye,fmt='.')
```

```python
# Peak Fit Functions
def a_peak(V, a):
    ''' See Eq. 6, lab guide '''
    C0, B, DE, E0, T = a
    B /= 1e22
    DE /= 1e10
    T /= 1e10
    return C0*(1-B/((E0*V/g.const.speed_of_light - DE)**2 + T**2))

def six_peaks(V, a):
    x = zeros(V.shape)
    for i in range(6):
        x += a_peak(V, array([a[i], a[i+6], a[i+12], a[-1], a[i+18]]))
    return x

def prep_for_fit(bundle, approx, misc=misc, vel_per_channel=cal.vel.vel_per_channel):
    DE = vel_per_channel[approx.peak_channel.astype(int)]/g.const.speed_of_light*misc.E0
    T = abs(vel_per_channel[(approx.peak_channel+approx.peak_fwhm/2).astype(int)]-
            vel_per_channel[(approx.peak_channel-approx.peak_fwhm/2).astype(int)])*misc.E0/(2*g.const.speed_of_light)
    B = -(approx.peak_mins/approx.peak_max_count-1)*T**2
    bundle.a0 = concatenate([approx.peak_max_count, B*1e22, DE*1e10, T*1e10, [misc.E0]], axis=1)

def run_fit(fe):
    fe.a, fe.aerr, fe.chisq, fe.yfit = j.levmar(fe.vel, fe.y, fe.ye, six_peaks, fe.a0, print_dev=True)[:-1]
    
def plot_fit(fe):
    print('$chi^sq$: {}'.format(fe.chisq))
    errorbar(fe.vel, fe.y, fe.ye, fmt='.')
    plot(fe.vel, fe.yfit, 'r')
    xlabel('Velocity [1]')
    ylabel('Counts [1]')
```

```python
# Manual Intervention for Approximate location of peaks
approx1 = Bundle()
approx1.peak_channel = array([276, 614, 948, 1210, 1578, 1960])
approx1.peak_max_count = array([110,110,110,110,110,110])/6
approx1.peak_mins = array([78,80,86,88,82,82]) - 5*int(mean(approx1.peak_max_count))
approx1.peak_fwhm = array([50,50,50,50,50,50])
prep_for_fit(fe1, approx1)

approx2 = Bundle()
approx2.peak_channel = array([276, 614, 948, 1210, 1578, 1960])
approx2.peak_max_count = array([80,80,80,80,80,80])/6
approx2.peak_mins = array([50,50,55,55,50,50]) - 5*int(mean(approx2.peak_max_count))
approx2.peak_fwhm = array([50,50,50,50,50,50])
prep_for_fit(fe2, approx2)

figure()
errorbar(fe1.vel, fe1.y, fe1.ye, fmt='.')
plot(cal.vel.vel_per_channel, six_peaks(cal.vel.vel_per_channel, fe1.a0), 'b')
errorbar(fe2.vel, fe2.y, fe2.ye, fmt='.')
plot(cal.vel.vel_per_channel, six_peaks(cal.vel.vel_per_channel, fe2.a0), 'g')
```

```python
# Conduct Six Peak Fit (slow)
run_fit(fe1)
run_fit(fe2)
```

```python
figure()
map(plot_fit, [fe1, fe2])
```
